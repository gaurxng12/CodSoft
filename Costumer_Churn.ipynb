```python
df = df.drop("CustomerId",axis=1)
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for column in df.columns:
    if df[column].dtype == np.number:
        continue
    df[column]=le.fit_transform(df[column])
```

```python
df
```

```python
x= df.drop("Exited",axis=1)
```

```python
y=df.Exited
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
x_feature=ss.fit_transform(x)
```

```python
x_feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_feature,y,test_size=0.2,random_state=1)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.fit(xtest,ytest)
```

```python
predict=lr.predict(xtest)
```

```python
lr.score(xtest,ytest)
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
lr1=GradientBoostingClassifier()
lr1.fit(xtrain,ytrain)
```

```python
lr1.score(xtrain,ytrain)
```

```python
predict=lr1.predict(xtest)
```

```python
lr1.score(xtest,ytest)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
lr2=RandomForestClassifier(n_estimators=30)
lr2.fit(xtrain,ytrain)
```

```python
lr2.score(xtrain,ytrain)
```

```python
predict=lr2.predict(xtest)
```

```python
lr2.score(xtest,ytest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,predict)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ytest,predict)
ax=plt.axes()
sns.heatmap(cm,annot=True,
           annot_kws={'size':10},
           ax=ax)
plt.title("Confusion_Matrix")
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,predict))
print(xtrain.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
from keras import utils
import seaborn
```

```python
model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(12,),activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,epochs=100)
```

```python
pred=model.predict(xtest)
```

```python
score = model.evaluate(xtest,ytest)
print("Loss: ",score[0])
print("Accuracy: ",score[1])
```
